```python
# importando o conjunto de dados de flores
import seaborn as sns

# salvando os dados em uma variável
iris = sns.load_dataset('iris')
```

```python
# apresentando os dados
iris
```

```python
# importanto a biblioteca para plotar os dados
import matplotlib.pyplot as plt
 
# plotando os dados
sns.FacetGrid(iris, 
              hue ="species",
              height = 6).map(plt.scatter,
                              'sepal_width',
                              'petal_width').add_legend()
```

```python
# verificando os titulos das colunas
iris.columns
```

```python
# separando as entradas 
X = iris.loc[:,['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# apresentando as entradas 
X
```

```python
# separando as saidas 
y = iris.loc[:,['species']]

# apresentando as entradas 
y
```

```python
# importando a biblioteca para separar os dados em treino e teste
from sklearn.model_selection import train_test_split

# dividindo o dataset em treinamento (70%) e teste (30$)%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# verificando o conjunto de teste
X_test
```

```python
# importando a biblioteca da árvore de decisão
from sklearn import tree

# instanciando a arvore de decisao
arvore = tree.DecisionTreeClassifier()

# pegando os parametros de treinamento
arvore.get_params(deep=True)
```

```python
# treinando a arvore de decisao
arvore.fit(X_train, y_train)
```

```python
# avaliando a classificação da arvore para o conjunto de teste
y_prev_test = arvore.predict(X_test)

# avaliando a classificacao
y_prev_test
```

```python
# importando a metrica de avaliação de desempenho
from sklearn.metrics import accuracy_score

# avaliando a taxa de acerto da arvore
accuracy_score(y_test, y_prev_test)
```
